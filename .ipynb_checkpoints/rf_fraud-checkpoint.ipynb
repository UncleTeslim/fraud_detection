```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical features
cat_features = ["gender", "category", "merchant", "job"]
encoder = LabelEncoder()
for col in cat_features:
    raw_data_no_info[col] = encoder.fit_transform(raw_data_no_info[col])
raw_data_no_info.head(5)
```

#### Feature Engineering

```python
#Relative transaction size
raw_data_no_info["amt_vs_avg_30d"] = raw_data_no_info["amt"] / (raw_data_no_info["customer_avg_amount_30_day"] + 1)
raw_data_no_info["amt_vs_avg_7d"] = raw_data_no_info["amt"] / (raw_data_no_info["customer_avg_amount_7_day"] + 1)
raw_data_no_info["amt_vs_avg_1d"] = raw_data_no_info["amt"] / (raw_data_no_info["customer_avg_amout_1_day"] + 1)

#Customer & Merchant Transaction Velocity
raw_data_no_info["customer_trans_rate_7d"] = raw_data_no_info["customer_num_trans_7_day"] / 7
raw_data_no_info["customer_trans_rate_30d"] = raw_data_no_info["customer_num_trans_30_day"] / 30
raw_data_no_info["merchant_trans_rate_7d"] = raw_data_no_info["merchant_num_trans_7_day"] / 7
raw_data_no_info["merchant_trans_rate_30d"] = raw_data_no_info["merchant_num_trans_30_day"] / 30

# Merchant Risk Indicators
raw_data_no_info["merchant_risk_change_7d_30d"] = raw_data_no_info["merchant_risk_7_day"] - raw_data_no_info["merchant_risk_30_day"]
raw_data_no_info["merchant_risk_change_30d_90d"] = raw_data_no_info["merchant_risk_30_day"] - raw_data_no_info["merchant_risk_90_day"]

#Transaction Time-Based Features
raw_data_no_info["hour"] = raw_data_no_info["trans_time"].str.split(":").str[0].astype(int)
raw_data_no_info["is_night"] = (raw_data_no_info["hour"] >= 22) | (raw_data_no_info["hour"] <= 6)
raw_data_no_info["is_weekend"] = raw_data_no_info["trans_date_is_weekend"]


raw_data_no_info = raw_data_no_info.drop(columns=["customer_avg_amout_1_day", "customer_avg_amount_7_day",
                                  "customer_avg_amount_30_day", "trans_date", "trans_time","trans_time_secs",
                                  "trans_time_hrs", "customer_num_trans_7_day", "customer_num_trans_30_day",
                                 "merchant_num_trans_7_day", "merchant_num_trans_30_day"])

raw_data_no_info.head(5)
```

```python
night_feature = ["is_night"]
encotder = LabelEncoder()
for col in night_feature:
    raw_data_no_info[col] = encoder.fit_transform(raw_data_no_info[col])
raw_data_no_info.head(5)
```

```python
raw_data_no_info.to_csv('clean_data.csv', index=False)
```

```python
# from sklearn.feature_selection import SelectFromModel
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# selector = SelectFromModel(rf, threshold=0.003)
# selector.fit(X, y)
selected_features = ['amt', 'amt_vs_avg_7d', 'amt_vs_avg_30d', 'amt_vs_avg_1d', 'is_night', 'is_weekend', 'hour', 'gender']



X_selected = featured_data[selected_features]
y = featured_data["is_fraud"]


X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

rf_final = RandomForestClassifier(n_estimators=200, max_depth=15, n_jobs=-1, random_state=42)
rf_final.fit(X_train_balanced, y_train_balanced)

# Predictions
y_pred = rf_final.predict(X_test)
y_prob = rf_final.predict_proba(X_test)[:, 1]

# Metrics
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_prob)
report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("F1-Score:", f1)
print("ROC AUC Score:", roc_auc)
print("Classification Report:\n", report)

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```
